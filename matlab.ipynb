```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title('Chart')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,y**2)
plt.subplot(2,2,4)
plt.plot(y,x**2)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.0,0.0,.5,.5])
plt.plot(x,y)
axis2 = fig.add_axes([0.54,0.0,.5,.5])
plt.plot(y,x)
```
