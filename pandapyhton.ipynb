```python
def extract_ep(txt):
    check = False
    data = ''
    for i in txt:
        if i == ')':
            check = False
            break
        if check == True:
            data += i
        if i == '(':
            check = True
    return data
```

```python
df['Episodes'] = df['Title'].apply(extract_ep)
```

```python
df['Episodes'] = df['Episodes'].str.replace(" eps","")
```

```python
def extract_time(txt):
    data =''
    for i in range(len(txt)):
        if txt[i] ==')':
            for j in range(i+1,i+20):
                data += txt[j]
            return data
```

```python
df['Total time'] = df['Title'].apply(extract_time)
```

```python
df
```

```python
from datetime import datetime
def calculate_months(date_range):
    try:
        start_str, end_str = date_range.split(" - ")

        start_date = datetime.strptime(start_str.strip(), "%b %Y")
        end_date = datetime.strptime(end_str.strip(), "%b %Y")

        total_months = (
            (end_date.year - start_date.year) * 12
            + (end_date.month - start_date.month)
        )

        return total_months+1
    except:
        return None
df["total_months"] = df["Total time"].apply(calculate_months)
```
